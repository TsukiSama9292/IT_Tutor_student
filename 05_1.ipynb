## ① Key-Value 概念

字典就像一本真正的字典：
- 你查「apple」（**key**）→ 得到「蘋果」（**value**）
- 你查「school」（**key**）→ 得到「學校」（**value**）

Python 字典就是一堆 **key: value** 配對的集合。

**語法**：
```python
# {key1: value1, key2: value2, ...}
student = {"name": "小明", "age": 16, "score": 85}
```

**重點規則**：
- key 必須是「不可變的」：字串、數字、tuple 可以；list 不行
- value 可以是任何型態（數字、字串、list、甚至另一個 dict）
- 同一個 key 不能重複（後來的會蓋掉前面的）

````python
# 建立字典
empty = {}
print("空字典:", empty)

student = {
    "name": "小明",
    "age": 16,
    "score": 85
}
print("學生資料:", student)

# 用 key 取值
print("姓名:", student["name"])
print("年齡:", student["age"])
print("成績:", student["score"])
````

**⚠️ KeyError 陷阱**：如果 key 不存在，用 `[]` 會報錯！
安全做法：使用 `.get()`

````python
# .get() 安全取值
print(student.get("height"))          # None（不會報錯）
print(student.get("height", "無資料"))  # 無資料（自訂預設值）

# 如果硬要用 [] 取不存在的 key 會怎樣？
# print(student["height"])  # 解除註解就會看到 KeyError
````

## ② 走訪字典：keys / values / items

這三種方法讓你遍歷字典的不同部分：

````python
student = {"name": "小明", "age": 16, "score": 85}

# 取得所有 keys
print("keys:", student.keys())
for k in student.keys():
    print("  key:", k)

# 取得所有 values
print("values:", student.values())
for v in student.values():
    print("  value:", v)

# 取得所有 items（key, value 成對）
print("items:", student.items())
for k, v in student.items():
    print(f"  {k} → {v}")
````

## ③ 修改字典：新增 / 更新 / 刪除

字典是可變的（mutable），可以隨時新增、更新、刪除資料。

````python
student = {"name": "小明", "age": 16, "score": 85}
print("原本:", student)

# 新增 key-value
student["height"] = 170
print("新增height:", student)

# 更新（同一個 key）
student["score"] = 95
print("更新score:", student)

# 刪除
del student["age"]
print("刪除age:", student)

# pop 取出並刪除
name = student.pop("name")
print("pop name:", name)
print("剩下:", student)
````

**注意**：
- 新增和更新用**同樣的語法**（`dict[key] = value`）—— key 不存在就是新增，已存在就是更新
- `del` 直接刪除，但 key 不存在會報錯
- `.pop(key, default)` 安全刪除並回傳值，可給預設值

## ④ 巢狀結構：List of Dicts

一個班級有很多學生，每個學生是一個 dict，全班就是一個 **list of dicts**。
這就像 Excel 表格：每一列是一個學生，每一欄是一個屬性。

````python
students = [
    {"name": "小明", "age": 16, "score": 85},
    {"name": "小華", "age": 17, "score": 92},
    {"name": "小美", "age": 16, "score": 78}
]

# 依序印出所有學生
for s in students:
    print(f"{s['name']}: {s['score']}分")

# 找出最高分的學生
top = students[0]
for s in students:
    if s["score"] > top["score"]:
        top = s
print(f"\n最高分: {top['name']} — {top['score']}分")
````

---
## ⑤ 巢狀結構：Dict of Dicts

用一個唯一的 ID（如學號）當 key，值則是一個 dict。適合「快速查找」。

````python
students = {
    "A001": {"name": "小明", "age": 16, "score": 85},
    "A002": {"name": "小華", "age": 17, "score": 92},
    "A003": {"name": "小美", "age": 16, "score": 78}
}

# 直接查學號
print("A002 姓名:", students["A002"]["name"])

# 走訪所有學生
for sid, info in students.items():
    print(f"{sid}: {info['name']} — {info['score']}分")
````
