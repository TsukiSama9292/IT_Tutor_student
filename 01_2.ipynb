## ① 認識 print() 與字串

`print()` 是用來在螢幕上顯示結果的函數。
字串（string）就是用 `"` 或 `'` 包起來的文字。

````python
print("Hello")          # 雙引號
print('Hello')          # 單引號也可以
print("Hello" + " " + "World")  # 字串拼接
print("Hello\nWorld")   # \n 換行
print("=" * 20)         # 字串乘法
print("Ha" * 5)
````

## ② 變數與資料型態

變數就像一個「貼了標籤的盒子」，可以用來存放資料。

````python
# 整數 (int)
age = 16
print("年齡：", age)
print("型態：", type(age))

# 浮點數 (float)
pi = 3.14159
print("\n圓周率：", pi)
print("型態：", type(pi))

# 字串 (str)
name = "Alex"
print("\n名字：", name)
print("型態：", type(name))

# 布林值 (bool)
is_student = True
print("\n是否為學生：", is_student)
print("型態：", type(is_student))
````

## ③ 用 input() 和電腦對話

`input()` 可以讓使用者輸入資料。
**注意**：`input()` 回傳的資料型態永遠是**字串**！

````python
name = input("請輸入你的名字：")
print("你好，" + name + "！歡迎來到 Python 世界！")
````

### 型態轉換

如果要把 input 拿到的文字轉成數字，需要 `int()` 或 `float()`。

````python
age = input("請輸入你的年齡：")
print("輸入的型態：", type(age))  # 字串

age = int(age)
print("轉換後的型態：", type(age))  # 整數
print("明年你將", age + 1, "歲")
````
