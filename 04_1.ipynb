## ① 建立串列

串列就像一個有順序的置物櫃——用中括號 `[]` 把資料包起來，資料之間用逗號隔開。

````python
fruits = ["蘋果", "香蕉", "橘子", "葡萄", "西瓜"]

print("fruits[0]:", fruits[0])   # 蘋果  ← 第 1 個
print("fruits[1]:", fruits[1])   # 香蕉  ← 第 2 個
print("fruits[3]:", fruits[3])   # 葡萄  ← 第 4 個
print("fruits[-1]:", fruits[-1]) # 西瓜  ← 倒數第 1 個
print("fruits[-2]:", fruits[-2]) # 葡萄  ← 倒數第 2 個
````

## ③ 切片 Slicing

一次拿一段連續的元素，用 `list[start:end]`。

**🔑 口訣**：含前不含後（包含 `start`，不包含 `end`）

````python
fruits = ["蘋果", "香蕉", "橘子", "葡萄", "西瓜"]

print("fruits[1:4]:", fruits[1:4])  # 從 1 到 3（不含 4）
print("fruits[:3]:", fruits[:3])   # 從頭到 2
print("fruits[2:]:", fruits[2:])   # 從 2 到尾
print("fruits[:]:", fruits[:])     # 全部複製一份
````

## ⑤ 串列方法（新增/刪除/排序）

### 新增：append / insert

````python
todo = ["買早餐", "寫作業"]

# append — 加在最後面
todo.append("練鋼琴")
print("append 後：", todo)

# insert — 插入到特定位置
todo.insert(1, "掃地")
print("insert 後：", todo)
````

### 刪除：remove / pop

````python
todo = ["買早餐", "掃地", "寫作業", "練鋼琴"]

# remove — 刪除指定值（只刪第一個符合的）
todo.remove("掃地")
print("remove 後：", todo)

# pop — 用索引刪除，會回傳被刪掉的值
removed = todo.pop(1)
print("pop 後：", todo)
print("完成了：", removed)

# pop() 不加索引 = 刪最後一個
last = todo.pop()
print("再次 pop 後：", todo)
print("最後一個：", last)
````

### 排序與反轉：sort / reverse

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
print("原始：", numbers)

numbers.sort()
print("sort 後：", numbers)

numbers.reverse()
print("reverse 後：", numbers)

# sort(reverse=True) 可以直接從大到小
numbers.sort(reverse=True)
print("sort(reverse=True) 後：", numbers)
````

## ⑥ 走訪串列（三種方法）

### 方法一：for item in list（直接取值）

````python
fruits = ["蘋果", "香蕉", "橘子", "葡萄"]

for fruit in fruits:
    print("我喜歡吃", fruit)
````

### 方法二：for i in range(len(list))（用索引）

````python
fruits = ["蘋果", "香蕉", "橘子", "葡萄"]

for i in range(len(fruits)):
    print(f"第 {i+1} 個水果是 {fruits[i]}")
````

### 方法三：enumerate（兩個都要！）

````python
fruits = ["蘋果", "香蕉", "橘子"]

for i, fruit in enumerate(fruits):
    print(f"索引 {i} → {fruit}")
````

## ⑧ 串列生成式 List Comprehension（加分！）

如果時間允許，來學一個 Python 獨有的酷東西——一行就搞定串列！

````python
# 傳統寫法
squares = []
for i in range(1, 6):
    squares.append(i ** 2)
print("傳統寫法：", squares)

# 串列生成式（一行搞定）
squares = [i ** 2 for i in range(1, 6)]
print("生成式：", squares)
````

````python
# 加上條件
even_squares = [i ** 2 for i in range(1, 11) if i % 2 == 0]
print("偶數的平方：", even_squares)
````
