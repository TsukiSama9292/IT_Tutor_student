## ② Linear Search（線性搜尋）

**概念**：從頭到尾，一個一個檢查。

**類比**：翻課本從第 1 頁開始，一頁一頁找關鍵字。

**回傳值**：找到 → 回傳索引；找不到 → 回傳 -1

````python
def linear_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

numbers = [5, 3, 8, 1, 9, 2]
print(linear_search(numbers, 8))   # 2
print(linear_search(numbers, 10))  # -1
````

## ③ Binary Search（二元搜尋法）

**⚠️ 前提**：資料必須是**已排序**的！

**概念**：每次和中間值比對，把搜尋範圍砍掉一半。

**類比**：電話簿找人名——翻到中間，看目標在前面還後面，每次排除一半。

```
資料： [1, 3, 5, 7, 9, 11, 13, 15]
要找： 11

Step 1: 中間 = 7 → 11 > 7 → 取右半
Step 2: 中間 = 11 → 找到了！只用 2 步
```

````python
def binary_search(data, target):
    left = 0
    right = len(data) - 1

    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

sorted_numbers = [1, 3, 5, 7, 9, 11, 13, 15]
print(binary_search(sorted_numbers, 11))  # 5
print(binary_search(sorted_numbers, 4))   # -1
````

## ④ 速度大對決：Linear vs Binary

用 `time.time()` 看看兩個演算法實際差多少！

````python
import time

big_list = list(range(1000000))
target = 999999

# 線性搜尋
start = time.time()
linear_search(big_list, target)
end = time.time()
print(f"線性搜尋：{end - start:.4f} 秒")

# 二元搜尋
start = time.time()
binary_search(big_list, target)
end = time.time()
print(f"二元搜尋：{end - start:.4f} 秒")
````

**時間複雜度直覺**：

| 演算法 | 100 萬筆資料的步數 | 口訣 |
|--------|-------------------|------|
| 線性搜尋 | ~1,000,000 步 | O(n) — 資料變 2 倍，時間變 2 倍 |
| 二元搜尋 | ~20 步 | O(log n) — 資料變 2 倍，時間多 1 步 |
| 氣泡排序 | ~500,000,000,000 步 | O(n²) — 資料變 2 倍，時間變 4 倍 |

## ⑤ Bubble Sort（氣泡排序法）

**概念**：從左到右，兩兩比較相鄰的元素，左邊 > 右邊就交換。
每一輪結束，最大的數字會「浮」到最右邊（像氣泡一樣）。

**類比**：整理解題順序——比較相鄰兩題，難的放後面。

**視覺化**：
```
第一輪：
[5, 3, 8, 1, 9, 2] → 5>3 交換 → [3, 5, 8, 1, 9, 2]
→ 5<8 不動 → [3, 5, 8, 1, 9, 2]
→ 8>1 交換 → [3, 5, 1, 8, 9, 2]
→ 9>2 交換 → [3, 5, 1, 8, 2, 9]  ✓ 9 就定位
```

````python
def bubble_sort(data):
    n = len(data)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

numbers = [5, 3, 8, 1, 9, 2]
print(bubble_sort(numbers))
````

**為什麼內層是 `range(n - 1 - i)`？**
- 每一輪結束後，最後 `i` 個元素已經排好了
- 所以不需要再和它們比較
- 避免不必要的運算！

### 練習 1：實作 Selection Sort（選擇排序法）

**概念**：每次從未排序的部分「選擇」最小的元素，放到已排序的最後面。

```
開始: [64, 34, 25, 12, 22, 11, 90]
第 1 輪: [11, 34, 25, 12, 22, 64, 90]  (11 和 64 交換)
第 2 輪: [11, 12, 25, 34, 22, 64, 90]  (12 和 34 交換)
...
```

```python
def selection_sort(data):
    n = len(data)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if data[j] < data[min_idx]:
                min_idx = j
        data[i], data[min_idx] = data[min_idx], data[i]
    return data
```
