### range() — 數字產生器

`range()` 可以產生一系列的數字：
| 寫法 | 產生的數字 |
|------|-----------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |

````python
# range(stop) — 0 到 stop-1
print("range(5):")
for i in range(5):
    print(i, end=" ")

print("\n")  # 換行

# range(start, stop)
print("range(2, 7):")
for i in range(2, 7):
    print(i, end=" ")
````

````python
# range(start, stop, step) — 有間隔
print("偶數：")
for i in range(0, 10, 2):
    print(i, end=" ")

print("\n")  # 換行

print("倒數：")
for i in range(5, 0, -1):
    print(i, end=" ")
````

### 迭代串列

for 迴圈也可以一個一個取出串列裡面的東西。

````python
fruits = ["蘋果", "香蕉", "橘子", "西瓜"]
for fruit in fruits:
    print("我喜歡吃", fruit)
````

````python
# 搭配 range() 取得索引
fruits = ["蘋果", "香蕉", "橘子", "西瓜"]
for i in range(len(fruits)):
    print("第", i + 1, "個水果是", fruits[i])
````

## ② while 迴圈：做到條件不成立為止

for 是「固定次數」的事；while 是「做到條件不成立為止」。
就像——「一直吃飯，直到吃飽為止」。

---
## ① for 迴圈：固定次數的重複

想像你要對全班 30 個同學說「你好」——你會寫 30 行程式嗎？
不用！用 `for` 迴圈一行就搞定。

````python
# while 條件:
#     要做的事

count = 1
while count <= 5:
    print("第", count, "次迴圈")
    count += 1  # 重要：更新條件！

print("迴圈結束！")
````

### ⚠️ 無限迴圈警告

如果忘記更新條件，電腦就會一直做、一直做⋯⋯

### break — 強制跳出迴圈

`break` 可以讓你在迴圈中間就跳出來。

````python
# break 範例
count = 1
while True:
    print("第", count, "次")
    if count >= 5:
        print("夠了，停！")
        break
    count += 1

print("跳出迴圈了！")
````

### while vs for

| for 迴圈 | while 迴圈 |
|----------|-----------|
| 知道要做幾次 | 不知道要做幾次 |
| `for i in range(5):` | `while 條件:` |
| 自動更新變數 | 要手動更新 |
| 適合走訪串列 | 適合「等到某件事發生」 |

## ③ 實作挑戰：猜數字遊戲 🎯

現在來寫一個經典遊戲——**終極密碼**！
電腦隨機出一個數字，你來猜，每次都會告訴你「太大」、「太小」或「猜中了！」

### 第一步：匯入 random 模組

`random` 是 Python 內建的亂數模組，可以用來產生隨機數字。

````python
import random

# random.randint(a, b) 會回傳 a 到 b 之間的隨機整數（包含 a 和 b）
secret = random.randint(1, 100)
print("隨機數字產生了！（但先不要讓玩家看到）")
# print(secret)  # 測試時可以取消註解來看答案
````

### 第二步：完整遊戲

用 `while` 迴圈 + `if/elif/else` 組合出完整的猜數字邏輯。

````python
import random

secret = random.randint(1, 100)
guess = None

print("=== 終極密碼猜數字 ===")
print("我已經選好一個 1~100 的數字了！")

while guess != secret:
    guess = int(input("請輸入你的猜測："))
    
    if guess > secret:
        print("⬇️ 太大囉！再小一點")
    elif guess < secret:
        print("⬆️ 太小囉！再大一點")
    else:
        print("🎉 恭喜你！猜中了！")
````

## ④ 巢狀迴圈

迴圈裡面還可以放迴圈，就像時鐘——時針轉一圈，分針轉 60 圈。
外層跑 1 次，內層就會跑完整的一輪。

### 九九乘法表

用巢狀迴圈來產出經典的九九乘法表！

````python
# 外層：被乘數 1~9
for i in range(1, 10):
    # 內層：乘數 1~9
    for j in range(1, 10):
        print(i, "x", j, "=", i * j, end="\t")
    print()  # 換行
````

### 圖形輸出

用迴圈來畫各種形狀！

````python
# 直角三角形
print("直角三角形：")
for i in range(1, 6):
    print("*" * i)
````

````python
# 倒三角形
print("倒三角形：")
for i in range(5, 0, -1):
    print("*" * i)
````

````python
# 用巢狀迴圈畫矩形
print("矩形：")
for i in range(3):        # 3 行
    for j in range(5):    # 5 列
        print("*", end="")
    print()  # 換行
````
