# Day 6 — 函式：寫一次，用一輩子

到目前為止，我們寫的程式都是「從頭到尾執行一次」。
但如果有一段邏輯（例如計算 BMI、檢查質數）要在不同地方重複使用呢？
複製貼上？萬一要修改怎麼辦？

這時候就需要 **函式（function）**——把程式碼包裝起來，取個名字，隨時呼叫！

---
## ① 為什麼需要函式？

**DRY 原則**：Don't Repeat Yourself（不要重複自己）

**函式的好處**：
- 重複使用：寫一次，用很多次
- 可讀性：給程式碼取名字，一看就知道在做什麼
- 維護性：要修改功能，只需要改一個地方
- 模組化：把大問題拆成小功能

**類比**：函式就像「食譜」——
- 食譜有名稱（函式名稱）
- 食譜需要食材（參數）
- 照食譜做出菜（執行）
- 最後端出菜（回傳值）

---
## ② 定義與呼叫基本函式

用 `def` 關鍵字來定義函式：

```python
def 函式名稱():
    # 要做的事情
```

定義好之後，用 `函式名稱()` 來呼叫。

````python
# 定義一個最基本的函式
def say_hello():
    print("Hello!")

# 呼叫函式
say_hello()
say_hello()  # 可以一直呼叫！
````

````python
# 函式可以做很多事
def greet():
    print("歡迎來到 Python 課堂！")
    print("今天我們要學函式！")

greet()
````

**注意**：呼叫函式時**一定要加括號 `()`**。
如果只寫 `say_hello` 不加 `()`，Python 不會報錯，
但不會執行函式——它只會回傳函式物件本身。

---
## ③ 參數（Parameters）

參數讓函式可以接受「輸入」，根據不同的輸入產生不同的輸出。

**類比**：食譜的食材——同樣的食譜，用不同的食材做出不同的菜。

````python
# 一個參數
def greet(name):
    print(f"你好，{name}！")

greet("小明")
greet("小華")
````

````python
# 多個參數
def introduce(name, age):
    print(f"我叫{name}，今年{age}歲")

introduce("小明", 16)
introduce("小華", 17)
````

**重點**：
- 參數（parameter）：定義函式時寫的「變數名稱」
- 引數（argument）：呼叫函式時傳入的「實際值」
- 參數的數量、順序要對應

## ④ return vs print（超級重要！）

這是初學者**最常搞混**的地方！

- `print()`：把東西「顯示」在螢幕上，看完就沒了
- `return`：把東西「回傳」給呼叫的人，可以**繼續使用**這個值

**類比**：
- `print` 就像廚師把菜端出來給你看一眼，然後就收走了
- `return` 就像廚師把菜端出來給你——你可以吃（繼續使用）

````python
# 只有 print — 不能拿結果來用
def add_print(a, b):
    print(a + b)

result = add_print(3, 5)  # 螢幕上會顯示 8
print(result)            # None — 因為沒有回傳值！
````

````python
# 有 return — 可以拿結果繼續用
def add_return(a, b):
    return a + b

result = add_return(3, 5)   # 不會顯示東西
print(result)               # 8
print(add_return(10, 20))   # 30
````

**為什麼 `return` 很重要？**
因為我們可以把函式的結果**組合**起來使用！

````python
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

# 組合函式！
# 先算 add(3,5)=8, add(2,4)=6, 再算 multiply(8,6)=48
result = multiply(add(3, 5), add(2, 4))
print(result)  # 48
````

## ⑤ 預設參數（Default Parameters）

有些參數大部分時候都是一樣的值，可以給它一個「預設值」。
呼叫時如果不傳這個參數，就會用預設值。

````python
def greet(name, greeting="你好"):
    print(f"{greeting}，{name}！")

greet("小明")           # 用預設值：你好，小明！
greet("小華", "哈囉")    # 自訂：哈囉，小華！
greet("小英", "Hello")   # Hello，小英！
````

````python
def create_user(name, age, city="未設定"):
    return {"name": name, "age": age, "city": city}

u1 = create_user("小明", 16)
u2 = create_user("小華", 17, "台北")
print(u1)
print(u2)
````

**⚠️ 規則**：有預設值的參數**必須放在沒有預設值的參數之後**
- `def func(a, b=1):` ✅ 正確
- `def func(a=1, b):` ❌ 錯誤！

## ⑥ 作用域（Scope）基礎

函式裡面的變數是「區域變數（local variable）」，
只在函式內部有效。函式外面的變數是「全域變數（global variable）」。

````python
# 全域變數
x = 10

def show_x():
    print(x)  # 可以讀取全域變數

show_x()  # 10
````

````python
# 同名變數 — 函式內的 x 是新的區域變數！
x = 10

def change_x():
    x = 20  # 這是「區域變數」，不是外面的 x
    print("函式內:", x)

change_x()       # 函式內: 20
print("函式外:", x)  # 函式外: 10（沒有被影響！）
````

````python
# 區域變數在外面無法存取
def my_func():
    y = 5
    print(y)

my_func()  # 5
# print(y)  # 解除註解會看到 NameError！
````

**原則**：
- 函式透過**參數**取得輸入
- 函式用 `return` 回傳結果
- 儘量不要依賴函式外部的變數（容易出錯、難以除錯）
