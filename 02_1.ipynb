## ① 比較運算子

要讓電腦做判斷，首先要讓電腦能「比較」兩個東西。

| 運算子 | 意思 | 範例 |
|--------|------|------|
| `==` | 等於 | `5 == 5` → True |
| `!=` | 不等於 | `5 != 3` → True |
| `>` | 大於 | `5 > 3` → True |
| `<` | 小於 | `5 < 3` → False |
| `>=` | 大於等於 | `5 >= 5` → True |
| `<=` | 小於等於 | `5 <= 3` → False |

## ② if 單一條件判斷

`if` 的中文意思就是「如果⋯⋯就⋯⋯」。

**注意**：
- `if` 後面要加冒號 `:`
- 條件成立的程式碼要**縮排**（Tab 或 4 空格）

````python
age = 18

if age >= 18:
    print("你已經成年了！")
    print("可以考駕照了！")

print("這行程式一定會執行")
````

**試試看**：把 `age` 改成 `15`，再執行一次。

## ③ if / else 雙向判斷

「如果怎樣就做 A，**否則**就做 B」

````python
age = 15

if age >= 18:
    print("你已經成年了！")
else:
    print("你還未成年喔！")
    print("還有", 18 - age, "年才成年")
````

### 即興練習：奇偶數判斷

````python
number = int(input("請輸入一個整數："))
if number % 2 == 0:
    print(number, "是偶數！")
else:
    print(number, "是奇數！")
````

## ④ if / elif / else 多重判斷

現實中不只有兩種情況。例如成績有 A、B、C、D、F 五個等級。
`elif` 就是 `else if` 的縮寫。

````python
score = int(input("請輸入你的成績："))

if score >= 90:
    print("等級：A")
elif score >= 80:
    print("等級：B")
elif score >= 70:
    print("等級：C")
elif score >= 60:
    print("等級：D")
else:
    print("等級：F")
````

## ⑤ 邏輯運算子 and / or / not

真實的決策很少只有一個條件。可以用 `and`（而且）、`or`（或者）、`not`（否定）來組合條件。

````python
age = 17
has_id = True

if age >= 18 and has_id:
    print("可以進入")
else:
    print("不能進入")
````

````python
is_weekend = False
is_holiday = True

if is_weekend or is_holiday:
    print("今天不用上課！")
else:
    print("今天要上課")
````
