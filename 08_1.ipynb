````python
# 最簡單的類別
class Dog:
    pass

# 建立實體
my_dog = Dog()
print(type(my_dog))
print(isinstance(my_dog, Dog))
````

````python
# 加入 __init__ 建構子
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

my_dog = Dog("來福", "黃金獵犬")
print(my_dog.name)
print(my_dog.breed)

your_dog = Dog("小黑", "柴犬")
print(your_dog.name)
print(your_dog.breed)
````

**🔑 重點**：
- `__init__` 在建立新物件時**自動執行**
- `self` 代表「這個物件自己」
- `self.name = name` → 把參數存到這個物件的屬性
- 每個實體擁有自己的屬性值（來福 vs 小黑）

## ② 方法：類別裡面的函式

類別不只可以存資料（屬性），還可以有「行為」——也就是**方法（method）**。
方法就是**屬於這個物件的函式**，可以存取和操作這個物件的屬性。

---
## ① 什麼是類別？

還記得 `type()` 嗎？
- `type("hello")` → `<class 'str'>` → `str` **就是一個類別**！
- `type(42)` → `<class 'int'>` → `int` **也是一個類別**！

**重點**：
- 類別（class）= 型態、藍圖、模具
- 實體（object/instance）= 實際的資料、產品
- `"hello"` 是 `str` 類別的一個實體；`42` 是 `int` 類別的一個實體
- 今天開始，我們要**自己定義類別**！

**類比**：
| 類別（Class） | 實體（Object） |
|---|---|
| 樂高模具 | 樂高積木 |
| 食譜 | 蛋糕 |
| 學生資料表格 | 小明填好的那一格 |

````python
class Student:
    def __init__(self, name, age, grades):
        self.name = name
        self.age = age
        self.grades = grades

    def introduce(self):
        return f"我叫{self.name}，今年{self.age}歲"

    def average(self):
        return sum(self.grades) / len(self.grades)

    def is_passing(self, passing=60):
        return self.average() >= passing


s1 = Student("小明", 16, [85, 92, 78])
print(s1.introduce())
print(s1.average())
print(s1.is_passing())
````

**🔑 重點**：
- 方法的第一個參數永遠是 `self`
- 呼叫時**不用傳** `self`，Python 會自動帶入
- 方法內可以用 `self.xxx()` 呼叫自己的其他方法

## ③ 封裝（Encapsulation）

**引導問題**：*「如果銀行帳戶的餘額任何人都可以直接修改，會發生什麼事？」*

我們需要有個機制來**保護資料**——這就是封裝的概念。

Python 用 **`_` 前綴** 來表示：「這是我內部的資料，請不要直接碰。」

````python
# 不加保護 — 不安全
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

account = BankAccount("小明", 1000)
account.balance = -999999  # 任何人都能改成負的！
````

````python
# 用 _ 表示私有 + getter/setter
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self._balance = balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            return f"存款成功，餘額：{self._balance}"
        return "存款金額必須為正數"

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            return f"提款成功，餘額：{self._balance}"
        return "餘額不足或金額無效"

    def check_balance(self):
        return f"您的餘額：{self._balance} 元"


account = BankAccount("小明", 1000)
print(account.deposit(500))
print(account.withdraw(200))
print(account.check_balance())
# 雖然還是可以改，但 _ 就是「請不要這樣做」的記號
````

### Getter / Setter
用方法來控制屬性的讀取和寫入，可以加入驗證邏輯。

````python
class Student:
    def __init__(self, name, age):
        self.name = name
        self._age = age

    def get_age(self):
        return self._age

    def set_age(self, age):
        if 0 < age < 150:
            self._age = age
        else:
            print("年齡不合理！")


s = Student("小明", 16)
print(s.get_age())
s.set_age(17)
print(s.get_age())
s.set_age(-5)
````
