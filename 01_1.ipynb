## Cell ② 圖形繪畫

Python 也可以用來畫圖！`matplotlib` 是 Python 最常用的繪圖工具。

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 360, 100)
y_sin = np.sin(np.deg2rad(x))
y_cos = np.cos(np.deg2rad(x))

plt.figure(figsize=(8, 4))
plt.plot(x, y_sin, label="sin(θ)", color="blue", linewidth=2)
plt.plot(x, y_cos, label="cos(θ)", color="red", linewidth=2)
plt.axhline(y=0, color="gray", linestyle="--", alpha=0.5)
plt.title("Sine & Cosine Waves")
plt.xlabel("Angle (deg)")
plt.ylabel("Value")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print("Python can draw beautiful function graphs too!")
```

**試試看**：把 `blue` 改成 `green`，把 `red` 改成 `purple`，看看顏色變化！

## Cell ③ 互動占卜

Python 可以跟你互動，根據輸入做出不同回應！

```python
import random

name = input("請輸入你的名字：")

luck_level = random.randint(1, 100)

if luck_level >= 80:
    fortune = "大吉！今天運氣爆棚，去買張彩券吧！"
    lucky_item = "四葉草"
elif luck_level >= 50:
    fortune = "中吉！平穩順利的一天，適合學習新事物！"
    lucky_item = "原子筆"
elif luck_level >= 20:
    fortune = "小吉！有些事情需要多一點耐心喔！"
    lucky_item = "橡皮擦"
else:
    fortune = "末吉⋯⋯沒關係，明天會更好！"
    lucky_item = "一包衛生紙"

print("=" * 40)
print(f"🔮 {name} 的今日運勢")
print("=" * 40)
print(f"運勢指數：{luck_level} / 100")
print(f"命運預言：{fortune}")
print(f"幸運物：{lucky_item} 🍀")
print("=" * 40)
```

**試試看**：多執行幾次，每次的結果都一樣嗎？為什麼？

## Cell ④ 資料分析

Python 很擅長處理數字資料——算平均、找最大最小，一秒解決！

```python
# 一組學生的考試分數
scores = [88, 75, 92, 63, 79, 95, 84, 71]

print("成績列表：", scores)
print("=" * 40)
print(f"📊 最高分：{max(scores)} 分")
print(f"📊 最低分：{min(scores)} 分")

total = sum(scores)
count = len(scores)
average = total / count
print(f"📊 全班平均：{average:.1f} 分")

# 找出及格（>= 60）的人數
passed = [s for s in scores if s >= 60]
print(f"📊 及格人數：{len(passed)} / {count} 人")
print("=" * 40)
```

**試試看**：把 scores 裡面的數字改成你自己的分數，看看結果有什麼變化？

## Cell ⑤ 自動化重複

電腦最厲害的就是——不用喊累，重複做一樣的事！
來看看用 `for` 迴圈可以排出什麼圖案。

```python
# 用星星排出金字塔
height = 5

print("✨ 星星金字塔 ✨")
print("=" * 20)

for i in range(1, height + 1):
    spaces = " " * (height - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)

print("=" * 20)
```

**試試看**：把 `height = 5` 改成 `height = 10`，金字塔會變怎樣？如果改成 3 呢？
